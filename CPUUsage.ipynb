```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('D:\\Users\\DELL\\Downloads\\DataScience-Python388888888888888\\cpu1.xlsx')
df.head()
```

```python
plt.plot(df['Time'],df['Memory'])
plt.axhline(y=2.7,color='r',linestyle='-')
plt.axhline(y=2.3,color='b',linestyle='-')
plt.xlabel('Time')
plt.ylabel('CPU utilization')
plt.show()
```

```python
plt.plot(df['Time'],df['CPU'])
plt.axhline(y=2.7,color='r',linestyle='-')
plt.axhline(y=df['CPU'].mean(),color='y',linestyle='-')
plt.axhline(y=2.3,color='b',linestyle='-')
plt.xlabel('Time')
plt.ylabel('CPU utilization')
plt.show()
```

```python
plt.plot(df['Time'],df['CPU'])
plt.axhline(y=2.7,color='b',linestyle='-')
plt.axhline(y=df['CPU'].mean(),color='y',linestyle='-')
plt.axhline(y=df['CPU'].median(),color='g',linestyle='-')
plt.axhline(y=2.3,color='b',linestyle='-')
plt.xlabel('Time')
plt.ylabel('CPU utilization')
plt.show()
```

```python
plt.plot(df['Time'],df['CPU'])
plt.axhline(y=2.7,color='r',linestyle='-')
plt.axhline(y=df['CPU'].mean(),color='y',linestyle='--')
plt.axhline(y=df['CPU'].median(),color='g',linestyle='--')
plt.axhline(y=2.3,color='r',linestyle='-')
plt.xlabel('Time')
plt.ylabel('CPU utilization')
plt.show()
```

```python
plt.plot(df['Time'],df['CPU'])
plt.axhline(y=2.7,color='r',linestyle='-')
plt.text(0,2.7,'Higher Threshold',rotation=0)
plt.axhline(y=df['CPU'].mean(),color='y',linestyle='--')
plt.axhline(y=df['CPU'].median(),color='g',linestyle='--')
plt.axhline(y=2.3,color='r',linestyle='-')
plt.text(0,2.3,'Idle State',rotation=0)
plt.xlabel('Time')
plt.ylabel('CPU utilization')
plt.show()
```

```python
plt.plot(df['Time'],df['CPU'])
plt.axhline(y=df['CPU'].mean(),color='b',linestyle='-')
plt.axhline(y=df['CPU'].quantile(1),color='r',linestyle='-')

plt.axhline(y=df['CPU'].quantile(0.05),color='r',linestyle='-')

plt.xlabel('Time')
plt.ylabel('CPU utilization')
plt.show()
```

```python
plt.plot(df['Time'],df['CPU'])
plt.axhline(y=df['CPU'].mean()+2*df['CPU'].std(),color='y',linestyle='--')
plt.axhline(y=df['CPU'].mean()-2*df['CPU'].std(),color='g',linestyle='--')
plt.xlabel('Time')
plt.ylabel('CPU utilization')
plt.show()
```
